The first step, of course, is to load the data. We will be using a dataset of shampoo sales over 3 years.

```python
import csv
import datetime as dt

with open("shampoo.csv", "r") as shamwow:
    data = list(csv.reader(shamwow))[1:]
    
sales = [float(point[1]) for point in data]
dates = [dt.datetime.strptime("199" + point[0],'%Y-%m').date() for point in data]

print(sales[0:5])
print(dates[0:5])
```

Let's plot the data in a plot that's as scattered as my brain while writing this tutorial:

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Plot with X-axis as dates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.scatter(dates, sales)
plt.gcf().autofmt_xdate()
```

Now let's perform the actual linear regression. We're fitting the data to a polynomial of degree 1, or in other words, a linear equation. Hopefully I have at least that many degrees by the time I leave college...

```python
import numpy as np

# Converting the datetime objects into an integer
counter = 1
numeric_months = []
for date in dates:
    numeric_months.append(counter)
    counter += 1
    
linear_regression = np.polyfit(numeric_months, sales, 1)
```

Remember like 5 years ago when you learned about slope-intercept form? Y'know, y = mx + b where m is the slope and b is the y-intercept? Don't ask me why I still remember that, but we're gonna use it now to plot the linear regression line on the plot:

```python
m, b = linear_regression

# Plot with X-axis as dates
points = plt.scatter(numeric_months, sales)
line = plt.axline((0, b), slope=m, color="red")
```
